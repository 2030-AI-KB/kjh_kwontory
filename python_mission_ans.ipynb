```python
# Missoin_1.  구구단프로그램을 작성해주세요. 단을 입력하면 입력받은 단까지 모두 출력해주세요.
# Mission_1-1. 단을 입력하면 그 단만 출력하는 프로그램

num = int(input("단을 입력하세요.\n(숫자만 입력, 예: 9)\n"))
print(str(num)+"단")
for i in range(1,10):
    print(num,'*',i,'=',i*num)
```

```python
# Mission_1-2. 단을 입력하면 그 단까지 모두 출력하는 프로그램
# 예) 9를 입력하면 2단부터 9단까지 출력함

num = int(input("단을 입력하세요.\n(숫자만 입력하세요. 예: 9)\n"))
num_tuple=tuple(range(2,num+1))

for i in num_tuple:
    print(str(i)+"단")
    for j in range(1,10):
        print(i,'*',j,'=',i*j)
```

```python
#Missoin_2.  n개의 정수를 입력 받아서 짝수와 홀수를 구분하여 저장해주세요. 
#           저장된 수의 총합이 얼마인지와 짝수의 합이 큰지 홀수의 합이 큰지를 알려주는 코드를 작성해주세요
#           사용자 입력 값이 중복되었을 경우 중복 값을 배제하지 않음

# 프로그램에 필요한 변수 생성
count = int(int(input("입력할 정수의 개수를 적으세요.\n(숫자만 입력하세요. 예: 9)\n")))
num_list = []
even = []
odd = []
even_sum = 0
odd_sum = 0

# 사용자 입력 받기
for i in range(1,count+1):
    num = int(input("정수를 입력하세요."+"("+str(i)+"/"+str(count)+")"))
    num_list.append(num)

# 짝수 홀수 구분해서 리스트 생성 
for j in range(0,len(num_list)):
    if num_list[j]%2==0:
        even.append(num_list[j])
    else:
        odd.append(num_list[j])
print("입력한 정수의 개수는 총 "+str(count)+"개이고, 그 중 짝수는 "+str(len(even))+"개, 홀수는 "+str(len(odd))+"개입니다.")

# 총합을 구하고 구한 총합 출력하기
for x in range(0,len(even)):
    even_sum +=even[x]
for y in range(0,len(odd)):
    odd_sum += odd[y]
print("짝수의 총합: "+str(even_sum)+"\n홀수의 총합: "+str(odd_sum))

# 합계 비교해서 출력하기
if even_sum > odd_sum:
    print("짝수의 총합이 더 큽니다.")
else:
    print("홀수의 총합이 더 큽니다.")
```

```python
# Mission_6. 이름과 점수를 입력하면 학점을 반환하여 저장하는 코드를 작성해주세요.

# 이름 입력받기
name = input("이름을 입력하세요. ")

while True:
    # 점수 입력받기
    score = int(input("점수를 입력하세요. "))

    # 학점 반환하여 저장(기준은 내맘대로)
    if score in range(0,101):
        if score > 90:
            grade = 'A'
        elif score > 80:
            grade = 'B'    
        elif score > 70:
            grade = 'C'
        elif score > 60:
            grade = 'D'
        else:
            grade = 'F'
        break
    
    else:
        print("성적 입력값이 정상 범위를 벗어났습니다. 다시 입력해주세요.")
    
# 잘 저장됐는지 확인
print("이름: "+name+"\n성적: "+str(score)+"\n학점: "+str(grade))
```

```python
# Mission_9. 가위,바위,보 게임을 만들어주세요.. (이긴 횟수를 알려주세요.)
# 사용자 입력이 원하는대로 들어오지 않았을 경우를 생각해서 코딩을 하고 싶은데 못하겠음
# 입력이 제대로 들어오는 경우만 생각하고 구현해봄

# Mission_9-1.

count = 0 # 총 게임 수(무승부는 카운트하지 않음)
win_count = 0 # player 1이 이긴 횟수

while True:
    player_1 = input("가위, 바위, 보! ")
    player_2 = input("가위, 바위, 보! ")
    
    if player_1 == player_2:
        print("Draw! Try again!")
        continue
    
    if player_1 == "가위":
        if player_2 == "바위":
            print("Winner: player 2")
            count += 1
        if player_2 == "보":
            print("Winner: player 1")
            count += 1
            win_count += 1
        
    if player_1 == "바위":
        if player_2 == "보":
            print("Winner: player 2")
            count += 1
        if player_2 == "가위":
            print("Winner: player 1")
            count += 1
            win_count += 1
        
        
    if player_1 == "보":
        if player_2 == "가위":
            print("Winner: player 2")
            count += 1
        if player_2 == "바위":
            print("Winner: player 1")
            count += 1
            win_count += 1
     
    # 게임 진행 여부, 결과 출력
    game_set = input("한판 더?(Y/N)")
    if game_set == 'N':
        if win_count > (count-win_count):
            print("Player 1 Win!")
            print("Total: "+str(count)+"\nWin count: "+str(win_count))
        if win_count == (count-win_count):
            print("Draw!")
        else:
            print("Player 2 Win!")
            print("Total: "+str(count)+"\nWin count: "+str(count-win_count))
        break
    elif game_set == 'Y':   
        continue
```

```python
# Mission_9-2. 가위바위보 게임, 튜플을 사용하고, 예외처리를 해봄

count = 0 # 총 게임 수(무승부는 카운트하지 않음)
win_count = 0 # player 1이 이긴 횟수

win = (('가위','보'),('바위','가위'),('보','바위'))
lose = (('가위','바위'),('바위','보'),('보','가위'))
draw =(('가위','가위'),('바위','바위'),('보','보'))

while True:
    play = (input("가위, 바위, 보! "),input("가위, 바위, 보! "))

    if play in draw:
        print("Draw! Try again!")
        continue
    
    elif play in win:
        print("Winner: player 1")
        count += 1
        win_count += 1
        
    elif play in lose:
        print("Winner: player 2")
        count += 1
    else: # 사용자 입력 오류
        print("Your input is wrong. Try again.")
        continue
     
    # 게임 진행 여부, 결과 출력
    game_set = input("한판 더?(Y/N)")

    if game_set == 'N':
        if win_count > (count-win_count):
            print("Player 1 Win!")
            print("Total: "+str(count)+"\nWin count: "+str(win_count))
        if win_count == (count-win_count):
            print("Draw!")
        else:
            print("Player 2 Win!")
            print("Total: "+str(count)+"\nWin count: "+str(count-win_count))
        break
    elif game_set == 'Y':   
        continue
    else:
        print('error')
```

```python
# Mission_10. 커피 자판기의 계산프로그램을 만들려고 합니다.(최대 15잔을 뽑을 수 있어요. 잔당 200원)
#             돈을 값을 입력 받으면 커피를 뽑을 수 있는 잔수와 잔액을 알려주시면 됩니다. 

money = int(input("예산을 입력하세요. "))
coffee = 200
max_coffee = 15

if money < (coffee*max_coffee):
    div=divmod(money, coffee)
    print(str(div[0])+"잔의 커피를 마실 수 있습니다. 잔액은 "+str(div[1])+"원입니다.")
else:
    print("15잔의 커피를 마실 수 있습니다. 잔액은 "+str(money-(coffee*max_coffee))+"원입니다.")

```
